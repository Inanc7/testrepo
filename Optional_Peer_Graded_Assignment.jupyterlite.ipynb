```python
#Is there a significant difference in median value of houses bounded 
#by the Charles river or not? (T-test for independent samples]

#H_0: u_1 = u_2 ("there is no difference in mean values of houses bounded by 
# Charles Rever")
#H_a: u_1 =! u_2 ("there is a difference in mean values of houses bounded by 
# Charles Rever")

scipy.stats.levene(boston_df[boston_df["CHAS"] == 0]["MEDV"],
                   boston_df[boston_df["CHAS"] == 1]["MEDV"],center = "mean")

#since the p value is less than 0.05 we assume inequality of the variance

scipy.stats.ttest_ind(boston_df[boston_df["CHAS"] == 0]["MEDV"],
                      boston_df[boston_df["CHAS"] == 1]["MEDV"],equal_var = False)

#Conclusion: Since the p-value is less than alpha value 0.05, 
#we reject the null hypothesis as there is enough that there 
#is a statistical difference in house prices by river and not

```

```python
#Is there a difference in Median values of houses (MEDV)
#for each proportion of owner occupied units built prior 
#to 1940 (AGE)? (ANOVA)

boston_df.loc[(boston_df["AGE"] <= 35), "AGE GROUP"] = "35 years and younger"
boston_df.loc[(boston_df["AGE"] > 40) & (boston_df["AGE"] < 70), "AGE GROUP"] = "between 35 and 70 years"
boston_df.loc[(boston_df["AGE"] >= 70), "AGE GROUP"] = "70 years and older"

#H_0: u_1 = u_2 ( there is no difference in Median values of houses (MEDV)
#for each proportion of owner occupied units built prior 
#to 1940 (AGE)

#H_a: u_1 =! u_2 ( there is a difference in Median values of houses (MEDV)
#for each proportion of owner occupied units built prior 
#to 1940 (AGE)

scipy.stats.levene(boston_df[boston_df["AGE GROUP"] == "35 years and younger"]["MEDV"],
                   boston_df[boston_df["AGE GROUP"] == "between 35 and 70 years"]["MEDV"],
                   boston_df[boston_df["AGE GROUP"] == "70 years and older"]["MEDV"],
                   center = "mean")

#since the p value is less than 0.05 we assume inequality of the variance
                
thirtyfive_lower = boston_df [boston_df["AGE GROUP"] == "35 years and younger"]["MEDV"]
thirtyfive_seventy = boston_df [boston_df["AGE GROUP"] == "between 35 and 70 years"]["MEDV"]
seventy_older = boston_df [boston_df["AGE GROUP"] == "70 years and older"]["MEDV"]

f_statistics, p_value = scipy.stats.f_oneway(thirtyfive_lower,thirtyfive_seventy,seventy_older)
print ("F_Statistics: {0}, P-Value: {1}".format (f_statistics,p_value))
      
#Since p value is less than 0.05 we reject the null hypotesis and 
# we accept that there is a difference in Median values of houses (MEDV)
#for each proportion of owner occupied units built prior 
#to 1940 (AGE)
```
